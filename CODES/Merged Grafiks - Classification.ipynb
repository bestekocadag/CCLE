```python
import matplotlib.pyplot as plt
import numpy as np

# Verilerin tanımlanması
categories = ['Malignant Melanoma - PD-0325901','Lymphoid Neoplasm - Nilotinib','Malignant Melanoma - AZD6244']
values1 = [0.98, 0.1, 0.27]
values2 = [0.45, 0.95, 0.53]
values3 = [0, 0, 0.80]

# Çubukların genişliği
bar_width = 0.2

# Çubukların pozisyonlarının ayarlanması
r1 = np.arange(len(categories))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]


# Bar plot oluşturma
plt.bar(r1, values1, color='green', width=bar_width, label='RF', capsize=5)
plt.bar(r2, values2, color='blue', width=bar_width, label='SVM', capsize=5)
plt.bar(r3, values3, color='red', width=bar_width, label='XGBoost', capsize=5)

# Başlık ve eksen etiketlerinin eklenmesi
plt.title('Prediction Power of ML Models Compared to Tissue and Drug')
plt.xlabel('Tissue - Drug')
plt.ylabel('R Square')

# X eksenine kategori isimlerinin eklenmesi ve döndürülmesi
plt.xticks([r + bar_width for r in range(len(categories))], categories, rotation=45, ha='right')

# Legend eklenmesi
plt.legend()

# Grafiğin gösterilmesi
plt.tight_layout()  # Bu, grafiğin elemanlarının düzgün yerleşmesini sağlar
os.chdir('C:\\Users\\MONSTER\\Desktop\\TEZ SON\\CCLE\\ML RESULT')
plt.savefig('Best Prediction for all Models - Regression.png')
plt.show()

```
