```r
Histologies <- list.dirs(full.name = FALSE, recursive = FALSE)
```

```r
Histologies
```

```r
#install.packages("plyr")
suppressPackageStartupMessages(library(plyr))
```

```r
Drug_list <- list()
Histology_list <- list()
for(histology in Histologies){
    setwd(paste0(output_dir,"/",histology))
    if(length(list.dirs(full.name = FALSE, recursive = FALSE)) > 5){
        Drugs <- list.dirs(full.name = FALSE, recursive = FALSE)
        for(drug in Drugs){
            setwd(paste0(output_dir,"/",histology,"/",drug))
            if(dir.exists("Random Forest")){
                setwd(paste0(output_dir,"/",histology,"/",drug,"/","Random Forest"))
                Drug_data_RF <- read.delim(paste(drug, "- Result_RF.tsv"))
                Drug_list[[drug]] <- Drug_data_RF
    
            }
            
        
        }
        
    Histology_list[[histology]] <- ldply(Drug_list, "rbind")
    Histology_list[[histology]] = Histology_list[[histology]][!duplicated(Histology_list[[histology]]),]    
    Histology_list[[histology]] <- subset(Histology_list[[histology]], select = -.id)    
    write.table(Histology_list[[histology]],
               paste0(output_dir,"/",histology,"/",histology," - RF_Result.tsv"),
               row.names = FALSE,
               quote = FALSE,
               sep = "\t")
                
    }
    
    
}
```

```r
RESULT <- ldply(Histology_list, "rbind")
RESULT = RESULT[!duplicated(RESULT),]
RESULT = na.omit(RESULT)
RESULT = subset(RESULT, select = -.id)

filter <- order(RESULT$R_Square_Test, decreasing = TRUE)
RESULT <- RESULT[filter,]
RESULT

write.table(RESULT,
               "C:/Users/MONSTER/Desktop/TEZ/CCLE/ML RESULT/ALL_RF_Result.tsv",
               row.names = FALSE,
               quote = FALSE,
               sep = "\t")
```
