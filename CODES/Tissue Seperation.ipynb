```r
merged_table <- merged_table[,c("Primary Cell Line Name","Histology","Drug","IC50","ActArea")]
colnames(merged_table) <- c("Cell_Line","Histology","Drug","IC50","Activity_Area")

head(merged_table, 3)
```

```r
histology_df <- list()
for(histology in histologies){
    setwd(ccle_output_dir)
    dir.create(histology , showWarnings = FALSE)
    setwd(histology)

    filter <- merged_table$Histology %in% histology
    histology_df[[histology]] <- merged_table[filter, ]

    write.table(histology_df[[histology]],
                paste(histology, "- Histology.tsv" , sep = " "),
                row.names = FALSE,
                quote = FALSE,
                sep = "\t")
    setwd("..")
    
    } 
```
